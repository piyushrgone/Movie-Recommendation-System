```python
movies = movies_data[['id','title','overview','genres','original_language','production_companies','tagline']]
movies.head()
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies.sample(5)
```

```python
movies['genres'].unique()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies.columns
```

```python
movies['genres'] = movies['genres'].apply(lambda x: x.split(','))
```

```python
movies['genres'] = movies['genres'].apply(collapse)
```

```python
movies['genres']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['original_language'] = movies['original_language'].apply(lambda x: x.split())
```

```python
movies['production_companies'] = movies['production_companies'].apply(lambda x: x.split())
```

```python
movies['tagline'] = movies['tagline'].apply(lambda x: x.split())
```

```python
movies.sample(5)
```

```python
movies['tags'] = movies['genres'] + movies['original_language'] + movies['overview'] + movies['production_companies'] + movies['tagline']
```

```python
df = movies[['id','title','tags']]
df.sample   (5)
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df.head()
```

### Stemming 


```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

stem(df['tags'][0])    
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

### Vectorizing & Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity # angular distance 
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector.reshape(-1,1))
similarity
```
